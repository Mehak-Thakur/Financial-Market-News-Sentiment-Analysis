Title: Financial Market News Sentiment Analysis

Objecive: The objective of Financial Market News Sentiment Analysis is to analyze the sentiment of financial news articles, headlines, or social media posts to assess their potential impact on stock prices, market trends, or investor behavior. By leveraging sentiment analysis, financial institutions, investors, and analysts can gain insights into the market's emotional state and make informed trading or investment decisions.

Dataset: A typical dataset for this task includes financial news headlines or articles, labeled with their corresponding sentiment scores (positive, negative, or neutral). The dataset may also include metadata such as timestamps, sources, and related stock ticker symbols.

Example Dataset Columns
Date: The publication date of the news.
Headline/News Text: The content or summary of the financial news.
Sentiment Label: Predefined sentiment categories (e.g., positive, negative, neutral).
Sentiment Score (optional): Numerical representation of sentiment (e.g., -1 for negative, 0 for neutral, +1 for positive).
Stock Ticker: The company or financial instrument associated with the news.
Market Impact (optional): Binary indicator of whether the news impacted the stock price.

```python
news = []
for row in range(0, len(df.index)):
  news.append(' '.join(str(x) for x in df.iloc[row, 2:27]))
```

```python
type(news)
```

```python
news[0]
```

```python
X = news
```

```python
type(X)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(lowercase = True, ngram_range = (1,1))
```

```python
X = cv.fit_transform(X)
```

```python
X.shape
```

```python
y= df['Label']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify = y, random_state=2529)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators= 200)
```

```python
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
